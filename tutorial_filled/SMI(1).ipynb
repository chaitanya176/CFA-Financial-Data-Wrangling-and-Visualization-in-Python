```python
import pandas as pd
import numpy as np
```

```python
data_frame = pd.read_csv('../data/intraday_30min_EURCHF=X.csv', names=['Date_Time', 'open','high', 'low','close', 'volume'],index_col=0, parse_dates=True)
```

```python
data_frame
```

```python
df1=data_frame.at_time('17:30')
```

```python
df2=data_frame.at_time('23:30')
```

```python
frames = [df1, df2]
```

```python
result = pd.concat(frames, axis=0, ignore_index=False)
```

```python
final=result.sort_values(by=['Date_Time'])
```

```python
final.head()

```

```python
import matplotlib.pyplot as plt
%matplotlib inline
ticker = 'CHFEUR'
# Plot the adjusted close price
final['close'].plot(figsize=(10, 7))
# Define the label for the title of the figure
plt.title(" Close Price of %s" % ticker, fontsize=16)
# Define the labels for x-axis and y-axis
plt.ylabel('Price', fontsize=14)
plt.xlabel('Year', fontsize=14)
# Plot the grid lines
plt.grid(which="major", color='k', linestyle='-.', linewidth=0.5)
# Show the plot
plt.show()
```

```python
#save dataframe in csv with date and time
with open("chfeur.csv", "+w") as file:
    final.to_csv(file, index=True, na_rep='NA')
```
